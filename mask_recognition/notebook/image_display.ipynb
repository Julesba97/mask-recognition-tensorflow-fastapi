```python
from pathlib import Path
import numpy as np
import os
import matplotlib.pyplot as plt

#os.chdir('../')
```

```python
prepared_train_path = Path("./mask_recognition/data/numpy_data/prepared_train_data.npz")
prepared_val_path = Path("./mask_recognition/data/numpy_data/prepared_train_data.npz")

prepared_train_data = np.load(prepared_train_path)
train_images = prepared_train_data["image"]
train_labels = prepared_train_data["label"]
```

## Exploration d'images

```python
N = len(train_images)
plt.figure(figsize=(20, 15))
for i in range(0, 15):
    w = np.random.randint(N)
    plt.subplot(3, 5, i+1)
    plt.imshow(train_images[w], cmap ="gray")
    plt.title(train_labels[w])
```
