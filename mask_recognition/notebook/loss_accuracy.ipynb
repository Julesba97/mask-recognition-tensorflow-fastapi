```python
import matplotlib.pyplot  as plt
from pathlib import Path
import pandas as pd
import os

os.chdir("../")
os.chdir("../")
os.getcwd()
```

```python
history_path = Path("./artefacts/history.csv")
history_df = pd.read_csv(history_path)
```

## Visualisation de la perte (loss) du modèle

```python
fig = plt.figure()
plt.plot(history_df["loss"], color="blue", label="loss")
plt.plot(history_df["val_loss"], color="orange", label="val_loss")
fig.suptitle("Loss", fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

## Visualisation de la précision (accuracy) du modèle

```python
fig = plt.figure()
plt.plot(history_df["accuracy"], color="blue", label="accuracy")
plt.plot(history_df["val_accuracy"], color="orange", label="val_accuracy")
fig.suptitle("Accuracy", fontsize=20)
plt.legend(loc="upper left")
plt.show()
```
